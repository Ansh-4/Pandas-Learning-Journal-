# Aggregation function 
reduces a set of values into a single summary value. 
## used to summarize and anaylze data
often used with the groupby() function


```python
df = pd.read_csv(r"C:\Users\Ryuk\Jupyter\Notes\importing\pokemon-150.csv")
print(df)
```

## to find mean of the numeric columns we can do the following:
using mean aggregate functions (only for numeric and for that alphabets we can pass an argument)

```python
print(df.mean(numeric_only=True)) #mean
```

```python
print(df.sum(numeric_only=True)) #sum/addition
```

```python
print(df.min(numeric_only=True)) #minimum
```

```python
print(df.max(numeric_only=True)) #maximum
```

```python
print(df.count()) #counting
```

## count won't include any null values thats why Type2 is 67 as many pokemons don't have a type2

# Groupby() --> a function
In pandas, groupby is a powerful method used to split data into groups, apply operations on each group, and then combine the results.

Think of it as:

Split → Apply → Combine

```python
group = df.groupby("Type1")
# group is an object hence when you apply an aggrgate function to the group object it will return a series for you to work with.
print(group["Height"].mean())
```

```python
print(group["Height"].sum())
```

```python
print(group["Height"].min())
```

```python
print(group["Height"].max())
```

```python
print(group["Height"].count())
```
