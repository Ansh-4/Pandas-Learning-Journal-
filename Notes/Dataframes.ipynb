```python
import pandas as pd
```

# DATAFRAMES
A Tabular Data Structure with rows and columns. Unlike series a Dataframe is 2-Dimensional.
Think of it like an excel spreadsheet.
A dataframe can be created as by dataframe constructer available in pandas.


```python
dict = {
    "Name":["Ansh","Ayush","Aryan","Shivam"],
    "Age":[20,19,20,21]
}
df = pd.DataFrame(dict)
print(df)
```

# Custom Labels
similar to series, DataFrame's default index value (i.e 0,1,2,3---so on) can be replaced by custom labels or indexes.

```python
df = pd.DataFrame(dict, index=["Employee-1","Employee-2","Employee-3","Employee-4"])
print(df)
```

# Loc Property
## Loc stands for location by label
Again similar to series loc can help us locate a specific data/value at a specific location

```python
print(df.loc["Employee-2"])
```

# iloc property
## gives location of the desired value by the index integer.
0 --> First Row

1 --> Second Row 

2 --> Third Row

3 --> Fourth Row


```python
print(df.iloc[0])
```

```python
print(df.iloc[2])
```

# Adding a new Column to the DataFrame


```python
df["Lang"]=["Python/Go","Javascript","Java+CSS","Java(springboot)"]
df["Role"]=["Data_Scientist","Backend_Dev","UI/UX","Full-Stack_dev"]
print(df)
```

# Adding a new Row to the DataFrame

```python
new_row1 = pd.DataFrame({"Name":"Unnati", "Age":18, "Lang":"Python/R", "Role":"Physicist"},
                        index=["Employee-5"])
df = pd.concat([df, new_row1]) #concatenation used instead of append because append is dated.
print(df)
```

```python
#Mutiple rows at once can be created as
new_rows= pd.DataFrame([{"Name":"Ashutosh", "Age":19, "Lang":"Rust/Python", "Role":"Cyber-security"},
                        {"Name":"Aditi", "Age":18, "Lang":"N/A", "Role":"HR"}],
                       index=["Employee-6","Employee-7"])
df=pd.concat([df, new_rows])
print(df)
```
