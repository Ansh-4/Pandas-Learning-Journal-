# SELECTION
In pandas, selection means choosing specific rows, columns, or parts of a DataFrame.
It is one of the most important skills when working with data.

```python
df = pd.read_csv(r"C:\Users\Ryuk\Jupyter\Notes\importing\pokemon-150.csv")
print(df)
```

## here we are only selecting the name column to be displayed. fun fact here our output is actually a lot of series
in order to print the entire column instead of first five and last five (trunkated Version) we can use string as such:

```python
print(df["Name"])
```

```python
print(df["Name"].to_string())
```

## Entire column of Names is printed

```python
print(df["Height"].to_string())
```

## All the heights of the Pokemons in meter
Now we can try to output mutiple specified columns at once


```python
print(df[["Name", "Height", "Weight"]])
```

## That was selection by columns now we can try selection by rows

```python
print(df)
```

Each of these rows has a label and the default behavior for pandas is to give each row a number for a label starting with 0. (ascending order)
# loc
## we can loc property to print the row itself

```python
print(df.loc[78])
```

```python
df = pd.read_csv(r"C:\Users\Ryuk\Jupyter\Notes\importing\pokemon-150.csv", index_col="Name")
print(df)
```

by using a paramter of index_col we have made the data easier to access as you probably wont remember the exact integer index of the pokemon but can easily remember the name itself

```python
print(df.loc["Pikachu"])
```

```python
print(df.loc["Charizard"])
```

## we also pass an argument to select only the columns and range of rows we would like to.....

```python
print(df.loc["Mewtwo",["Height","Weight"]])
```

```python
print(df.loc["Charmander":"Kakuna",["Height","Weight"]].to_string())
```

## Integer Based indexing --> iloc property

```python
print(df.iloc[0:11])
```
