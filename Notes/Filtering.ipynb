# Filtering
Filtering means selecting only the rows that meet a specific condition.

You keep the rows you want â†’ and remove the rest. (Keeping the rows that match the given condition)

```python
df=pd.read_csv(r"C:\Users\Ryuk\Jupyter\Notes\importing\pokemon-150.csv")
```

```python
print(df)
```

filtering

```python
tall_pokemon = df[df["Height"]>=2]
print(tall_pokemon)
```

## right here we have filtered our dataframe to pokemons whose height is greater than or equal to 2 meters.

```python
# now for weight
heavy_pokemon = df[df["Weight"]>100]
print(heavy_pokemon)
```

## Filtered out Dataframe containing only the pokemon whose weight is above 100 kilograms.

```python
#Return Pokemon 
legendary = df[df["Legendary"]==True] # equal to one if its legendary and equal to zero if its not basically true and false (BOOLEAN)
print(legendary)
```

## Filtered out Dataframe containing only Legendary Pokemons.
# Now lets try some operators

```python
water_pokemon = df[(df["Type1"] == "Water") | 
                    (df["Type2"]=="Water")]
print(water_pokemon)
# or operator
```

```python
ff_pokemon= df[(df["Type1"]=="Fire") &
                (df["Type2"]=="Flying")]
print(ff_pokemon)
# and operator
```
